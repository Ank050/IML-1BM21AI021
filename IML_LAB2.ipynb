```python
import pandas as pd
import copy

def is_consistent(hypothesis, example):
    for i in range(len(hypothesis)):
        if example[i] != '?' and hypothesis[i] != example[i]:
            return False
    return True

def minimal_generalizations(s, d):
    generalizations = []
    for i in range(len(s)):
        if s[i] != d[i]:
            gen = list(s)
            gen[i] = '?'
            generalizations.append(gen)
    return generalizations

def minimal_specializations(g, d, attributes):
    specializations = []
    for i in range(len(g)):
        if g[i] == '?':
            for val in attributes[i]:
                if val != d[i]:
                    spec = list(g)
                    spec[i] = val
                    specializations.append(spec)
    return specializations

def candidate_elimination(examples):
    attributes = [["some", "many"], ["small", "big", "medium"], ["no"], ["affordable", "expensive"], ["one", "many", "few"]]

    G = [list('?' * (len(examples[0]) - 1))]
    S = [list('?' * (len(examples[0]) - 1))]

    for example in examples:
        if example[-1] == 'yes':
            G_copy = copy.deepcopy(G)
            to_remove_S = []
            for outer_g in G_copy:
                if not is_consistent(outer_g, example[:-1]):
                    G.remove(outer_g)
                    to_remove_S += [s for s in S if not is_consistent(s, example[:-1])]
            S = [s for s in S if s not in to_remove_S]
            S += [gen for s in to_remove_S for gen in minimal_generalizations(s, example[:-1])]
            S = [h for h in S if any(is_consistent(g, h) for g in G)]
            S = [h for h in S if not any(is_consistent(h, h_dash) and h != h_dash for h_dash in S)]

        else:
            S_copy = copy.deepcopy(S)
            to_remove_G = []
            for outer_s in S_copy:
                if is_consistent(outer_s, example[:-1]):
                    S.remove(outer_s)
                    to_remove_G += [g for g in G if is_consistent(g, example[:-1])]
            G = [g for g in G if g not in to_remove_G]
            G += [spec for outer_g in to_remove_G for spec in minimal_specializations(outer_g, example[:-1], attributes)]
            G = [h for h in G if any(is_consistent(h_dash, outer_g) for h_dash in S)]
            G = [h for h in G if not any(is_consistent(outer_g, g_dash) and outer_g != g_dash for g_dash in G)]

    return S, G

# Your provided dataset
data = [
    ["some", "small", "no", "affordable", "one", "no"],
    ["many", "big", "no", "expensive", "many", "yes"],
    ["many", "medium", "no", "expensive", "few", "yes"],
    ["many", "small", "no", "expensive", "many", "yes"]
]

S_hypotheses, G_hypotheses = candidate_elimination(data)
print("Final hypothesis for S:", S_hypotheses)
print("Final hypothesis for G:", G_hypotheses)
```
